### 1.1 分词的概念

分词是文本标记化（tokenization）的一种方法，用来将文本切分成词语粒度的token。英文分词工作比较简单，只要以空格为界将句子拆开即可，而中文分词需要根据语义及实体的识别，难度较大。

- 正向最大匹配法：从前往后取词，直到命中或剩下一个单字。
- 逆向最大匹配法：从后往前取词，其他逻辑和正向相同。
- 双向最大匹配法：两种算法都切一遍，然后根据大颗粒度词越多越好，非词典词和单字词越少越好的原则，选取其中一种分词结果输出。

## 2.1 语言模型中unigram、bigram、trigram的概念

首先解释一般意义下的N-gram模型。该模型基于这样一种假设，第N个词的出现只与前面N-1个词相关，而与其它任何词都不相关，整句的概率就是各个词出现概率的乘积。

根据这个概念，unigram、bigram、trigram分别对应一元、二元、三元语言模型。

```
import re

sentences = re.split('。|！|\!|\.|？|\?', document)
sentences
```
